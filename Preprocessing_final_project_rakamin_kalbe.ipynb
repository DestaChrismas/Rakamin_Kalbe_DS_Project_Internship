## Cleaning Data

setelah melakukan assesing pada data diapat ketidak sesuaian data
1. pada data customer terdapat type data yang tidak sesuai untuk colom "Income" yaitu memiliki type data "object" yang seharusnya bertipe data "float" ,pada colom "CustomerID" memiliki type data "int" akan dibuah menjadi "object/str" dan pada data customer juga terdapat nilai NaN sebanyak 3 record pada colom "Marital Status"

2. pada data Store terdapat type daa yang tidak sesuai untuk colom "storeID" yaitu Int64 ,karena colom ini tidak digunakan untuk perhitungan jadi colom ini diubah type datanya menjadi "object" agar memungkinkan  untuk mempertahankan data "storeID" dalam bentuk teks, yang bisa lebih fleksibel dalam hal penyimpanan, pemrosesan dan dapat menyimpan karakteristik unik toko dalam "storeID," seperti prefiks atau karakter khusus.

3. pada data Transaction akan dilakukan perubahan data pada beberapa colom,seperti:
	1. CustomerID (int64) menjadi CustomerID (Object)
	2. Date (Object) menjadi Date (Datetime)
	3. StoreID (Int64) menjadi (Object)

```python
#1
# Merubah type data
# Konversi kolom 'Income' menjadi string
customers['Income'] = customers['Income'].astype(str)

# Ganti koma dengan titik dalam kolom 'Income'
customers['Income'] = customers['Income'].str.replace(',', '.')

# Mengubah tipe data kolom 'Income' menjadi float
customers['Income'] = customers['Income'].astype(float)

# Mengubah tipe data kolom 'Income' menjadi float
customers['CustomerID'] = customers['CustomerID'].astype(str)
```

```python
# Menghapus baris dengan nilai NaN di kolom 'Marital Status'
customers = customers.dropna(subset=['Marital Status'])
```

```python
# Cek perubahan Data Customer
customers.info()
customers.isna().sum()
```

```python
#2

# Mengubah tipe data kolom 'Store' menjadi Int64
Store['StoreID'] = Store['StoreID'].astype(str)
```

```python
# Cek perubahan Data Store
Store.info()
```

```python
#3
# Mengubah tipe data kolom 'CustomerID' menjadi object
Transaction['CustomerID'] = Transaction['CustomerID'].astype(str)
# Mengubah tipe data kolom 'Date' menjadi datetime
Transaction['Date'] = pd.to_datetime(Transaction['Date'])
# Mengubah tipe data kolom 'StoreID' menjadi object
Transaction['StoreID'] = Transaction['StoreID'].astype(str)
```

```python
# Cek perubahan Data Transaction
Transaction.info()
```

### Marge Data & save Data for SQL & Dasbaoard

```python
# Gabung Data

gabung1 = pd.merge(Transaction, customers, on='CustomerID')
gabung2 = pd.merge(gabung1, Product, on='ProductID')
gabung3 = pd.merge(gabung2, Store, on='StoreID')
allData = gabung3
```

```python
#hasil gabungan data
allData.head()
```

```python
#save Data
allData.to_csv('AllData.csv', index=False)

customers.to_csv('Datacustomers.csv', index=False)
Product.to_csv('DataProduct.csv', index=False)
Store.to_csv('DataStore.csv', index=False)
Transaction.to_csv('DataTransaction.csv', index=False)
```
